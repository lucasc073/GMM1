```python
import matplotlib.pyplot as plt
import numpy as np
import random
xnoisy = []
y= []
rate = []
rates = list(range(1, 10))
for i in range(0, 1000):
    div = random.choice(rates)
    rate += [div]
    xtemp = list(range(0, 100, div))
    ytemp = []
    for j in range(len(xtemp)):
        ytemp += [-2*(xtemp[j]-50)**2 + 2000 + np.random.normal(0, 100)]
        xtemp[j] = j + np.random.normal(0, 5)
    xnoisy += [xtemp]
    y += [ytemp]
    if(i<100):
        for j in range(len(xtemp)):
            if(j==(int)((len(xtemp)-1)/2)): plt.plot(xnoisy[i][j],y[i][j], 'o', markeredgecolor="green", markerfacecolor="green")
            elif(j<(len(xtemp)-1)/2): plt.plot(xnoisy[i][j],y[i][j], 'o', markeredgecolor="blue", markerfacecolor="blue")
            elif(j>(len(xtemp)-1)/2): plt.plot(xnoisy[i][j],y[i][j], 'o', markeredgecolor="red", markerfacecolor="red")
        plt.show()
```

```python
averager = 0
for i in range(100):
    averager += rate[i]
averager /= 100
print(averager)
print(100/averager)
```

```python
import random as random
time = 0
avgtime = 0
state = 0
for i in range(100):
    while(state<19):
        if random.random()<0.39:
            state += 1
        time +=1
    avgtime += time
    time = 0
    state = 0
avgtime /= 100
print(avgtime)
```

```python
averagey = [0] * 100
averagex = [0] * 100
count = [0] * 100

for i in range(1000):
    for j in range(0, len(y[i])):
        count[j*rate[i]] += 1
        averagey[j*rate[i]] += y[i][j]
        averagex[j*rate[i]] += xnoisy[i][j]*rate[i]
        
for i in range(100):
    averagey[i] /= count[i]
    averagex[i] /= count[i]
for i in range(100):
        if(i==50): plt.plot(averagex[i],averagey[i], 'o', markeredgecolor="green", markerfacecolor="green")
        elif(i<50): plt.plot(averagex[i],averagey[i], 'o', markeredgecolor="blue", markerfacecolor="blue")
        elif(i>50): plt.plot(averagex[i],averagey[i], 'o', markeredgecolor="red", markerfacecolor="red")
```
