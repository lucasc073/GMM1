```python
import matplotlib.pyplot as plt
import numpy as np
import random
xnoisy = []
y= []
rate = []
rates = list(range(1, 10))
for j in range(0, 100):
    div = random.choice(rates)
    rate += [div]
    xtemp = list(range(0, 100, div))
    ytemp = []
    for i in range(len(xtemp)):
        ytemp += [-2*(xtemp[i]-50)**2 + 2000 + np.random.normal(0, 100)]
        xtemp[i] = i + np.random.normal(0, 5)
    xnoisy += [xtemp]
    y += [ytemp]
    plt.plot(xnoisy[j],y[j])
    plt.show()
```

```python
print(y)
```

```python
averager = 0
for i in range(100):
    averager += rate[i]
averager /= 100
averager = 100/averager
print(averager)
```

```python
import random as random
time = 0
avgtime = 0
state = 0
for i in range(100):
    while(state<19):
        if random.random()<0.39:
            state += 1
        time +=1
    avgtime += time
    time = 0
    state = 0
avgtime /= 100
print(avgtime)

    
```
